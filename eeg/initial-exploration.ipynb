```python
timing = pd.read_csv("../eeg_tests/timing.csv")
```

# 3-BACK

```python
p25_dat, epochs, events = get_user_data('P25', timing, targets = ["three-back"], tmin = 0, tmax = .7)
```

```python
ica = ICA(n_components = .99, method='fastica')
ica.fit(epochs)
ica.plot_components()
```

```python
ica_epochs = ica.get_sources(epochs["target"])
_ = ica_epochs.average([5]).plot()
_ = ica_epochs.average([6]).plot()
```

```python
ica_epochs = ica.get_sources(epochs["not-target"])
_ = ica_epochs.average([5]).plot()
_ = ica_epochs.average([6]).plot()
```
